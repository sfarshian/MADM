```python
# Library Imports
import pandas as pd
import math
```

```python
# Normalized Matrix
# Normalize the matrix using the Eucledian method

N = pd.DataFrame(
        {
            "X1": [0.843, 0.337, 0.421],
            "X2": [0.402, 0.562, 0.723],
            "X3": [0.839, 0.466, 0.28],
            "X4": [0.509, 0.53, 0.678],
            "X5": [0.13, 0.39, 0.911],
        },
        index=["A1", "A2", "A3"],
    )

N
```

```python
# Weights
weights = pd.DataFrame([[0.179, 0.062, 0.211, 0.017, 0.531]],
                              columns=N.columns)
weights
```

```python
# Criteria Types
criteria_types = pd.DataFrame([["-", "+", "+", "+", "+"]],
                              columns=N.columns)

criteria_types
```

```python
# Weighted Matirx
V = N.copy()

for attr in V.columns:
  V[attr] *= weights.loc[0, attr]

V.round(3)
```

```python
# Ideal Positive Solution
a_plus = pd.DataFrame(0.0, index=[0], columns=V.columns)
# Ideal Negative Solution
a_minus = pd.DataFrame(0.0, index=[0], columns=V.columns)
for attr in V.columns:
    criteria_type = criteria_types.loc[0, attr]

    if criteria_type == "+":
        a_plus.loc[0, attr] = V[attr].round(3).max()
        a_minus.loc[0, attr] =V[attr].round(3).min()
    else:
        a_plus.loc[0, attr] = V[attr].round(3).min()
        a_minus.loc[0, attr] =V[attr].round(3).max()


a_plus
```

```python
# relative_distance
distance = pd.DataFrame(0.0, index=V.index, columns=["+", "-"])

for alternative in V.index:
  distance.loc[alternative, "+"] = round(math.sqrt(((V.loc[alternative] - a_plus.iloc[0]) ** 2).sum()), 3)
  distance.loc[alternative, "-"] = round(math.sqrt(((V.loc[alternative] - a_minus.iloc[0]) ** 2).sum()), 3)

distance
```

```python
# closeness
distance["Cl"] = (distance["-"] / (distance["-"] + distance["+"])).round(3)
distance
```

```python
# Ranking
distance.sort_values(by="Cl", ascending=False)
```
